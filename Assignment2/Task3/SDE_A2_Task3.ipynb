# **Running the Google API on provided images.**
## 1. Get the images - Images were uploaded to the site https://jay0774.github.io/IMAGES/ more images can be uploaded to the same site.

```python
url = "https://jay0774.github.io/IMAGES/"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data,'html.parser')
image_list = soup.find_all('img')
print("Images uploaded to the sites are:\n")
for i in range(len(image_list)):
  print("Image Number",i+1," has url ",url+image_list[i]['src'])
```

## 2. Using the API to label the image.

```python
for i in image_list[:3]:
  # creating the client for Image Annotations
  client = vision.ImageAnnotatorClient()
  url_base = url
  image_location = i['src']
  # Getting the source of the image
  image_url = url_base + image_location
  # requesting the image
  r = requests.get(image_url,stream=True)
  img = vision.Image()
  img.source.image_uri = image_url

  # getting the image for detecting the logo and face locally using url
  urllib.request.urlretrieve(image_url,"i.jpg")
  im = mpimg.imread('i.jpg')


  # getting the label detection
  response_client = client.label_detection(image=img)
  # getting the labels given the detector 
  labels = response_client.label_annotations
  # printing the labels description of image 
  if r.status_code == 200:
    print("Label and Conficence for"+" "+image_url,'\n')
    for l in labels:
      print(l.description," has a score of ",l.score*100)
  print('\n')

  # getting the logo detection
  response_client = client.logo_detection(image=img)
  # getting the labels given the detector 
  labels = response_client.logo_annotations
  # showing the logo and the boundary ploygon  
  if r.status_code == 200:
    print("Logo and Conficence for"+" "+image_url,'\n')
    for l in labels:
      print(l.description," has a score of ",l.score*100)
    if l.bounding_poly:
      # getting the statring point of boundary polygon
      x1,y1 = l.bounding_poly.vertices[0].x,l.bounding_poly.vertices[0].y
      # getting the ending point of boundary polygon
      x2,y2 = l.bounding_poly.vertices[2].x,l.bounding_poly.vertices[2].y
      # drawing the boundary polygon using cv2 library
      cv2.rectangle(im, (x1,y1), (x2,y2),(0,255,0), 2)
      # showing the image using the matplotlib bibrary
      imgplot = plt.imshow(im)
      plt.show()
  print('\n')

  # getting the face detection
  response_client = client.face_detection(image=img)
  # getting the labels given the detector 
  labels = response_client.face_annotations
  # showing the faces and the boundary ploygon   
  if r.status_code == 200:
    print("Face and Conficence for"+" "+image_url,'\n')
    # for each detected face drawing the boundary rectangle.
    for l in labels:
      if l.bounding_poly:
        print("For a new face recognised")
        # getting the statring point of boundary polygon
        x1,y1 = l.bounding_poly.vertices[0].x,l.bounding_poly.vertices[0].y
        # getting the ending point of boundary polygon
        x2,y2 = l.bounding_poly.vertices[2].x,l.bounding_poly.vertices[2].y
        # drawing the boundary polygon using cv2 library
        cv2.rectangle(im, (x1,y1), (x2,y2),(0,255,0), 2)
        # showing the image using the matplotlib bibrary
        imgplot = plt.imshow(im)
        plt.show()
  print('\n\n')

```
